```
import numpy as np

a = np.array([0, 1])
b = np.array([1, 0])

a + b
```

```
from numpy import sqrt

sqrt(a[0] ** 2 + a[1] ** 2)
```

```
from numpy.linalg import norm
print(norm(a + b))
```

```
from numpy import dot
from numpy.linalg import norm

dot(a, b) / ( norm(a) * norm(b) )
```

```
def cosine(v1, v2):
  val = dot(v1, v2) / ( norm(v1) * norm(v2) )
  return val if np.isnan(val) == False else 0
```

# Praca na przykładowych danych (golf)


```
# uzywamy pandas'a do skadowania danych w formie tabelarycznej

import pandas as pd

golf_df = pd.read_csv('dane_golf.csv', sep =';')
golf_df
```

```
from sklearn.preprocessing import LabelEncoder

# tworzymy encoder do atrybutu 'Outlook' i transformujemy atrybut
outlook_encoder = LabelEncoder()
outlook_attribute = outlook_encoder.fit_transform(golf_df['Outlook'])

# tworzymy encoder do atrybutu 'Temperature' i transformujemy atrybut
temperature_encoder = LabelEncoder()
temperature_attribute = temperature_encoder.fit_transform(golf_df['Temperature'])

# tworzymy encoder do atrybutu 'Humidity' i transformujemy atrybut
humidity_encoder = LabelEncoder()
humidity_attribute = temperature_encoder.fit_transform(golf_df['Humidity'])

# tworzymy encoder do atrybutu 'Windy' i transformujemy atrybut
windy_encoder = LabelEncoder()
windy_attribute = windy_encoder.fit_transform(golf_df['Windy'])

# tworzymy encoder do atrybutu 'Play Golf' i transformujemy atrybut
play_encoder = LabelEncoder()
play_attribute = play_encoder.fit_transform(golf_df['Play Golf'])
```

```
# budujemy nowy dataframe na bazie wyników transformacji

golf_df_transformed = pd.DataFrame()
golf_df_transformed['Outlook'] = outlook_attribute
golf_df_transformed['Temperature'] = temperature_attribute
golf_df_transformed['Humidity'] = humidity_attribute
golf_df_transformed['Windy'] = windy_attribute
golf_df_transformed['Play Golf'] = play_attribute

golf_df_transformed
```

```
X_train = golf_df_transformed[['Outlook', 'Temperature', 'Humidity', 'Windy']].to_numpy()
y_train = np.squeeze(golf_df_transformed[['Play Golf']].to_numpy(), axis=1)
```

```
from sklearn.naive_bayes import CategoricalNB

# tworzymy model oparty o kalstyfikator bayesa (dla danych kategorycznych)
# nie ma mozliwosci zrobienia tego dla danych tekstowych
# https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.CategoricalNB.html

golf_model = CategoricalNB().fit(X_train, y_train)

# sprawdzamy jaki procent trafilismy dobrze
predicted = golf_model.predict(X_train)
print(np.mean(predicted == y_train))
```

```
# możemy też wykorzystać gotowy model do sprawdzania nowych przypadków
X_test = np.array([[0, 0, 0, 0]])

golf_model.predict_proba(X_test)
```

# Cztery dokumenty

```
d1 = "NLP – w kontekście nowych technologii oznacza przetwarzanie języka naturalnego. Programy komputerowe rozumieją tylko matematyczne, ustrukturyzowane dane. Dlatego informacje, które ten program ma interpretować, musimy przeformatować w coś, co zrozumie. Natural Language Processing zajmuje się dekodowaniem tekstu i głosu człowieka oraz przetwarzaniem go na dane zrozumiałe dla komputera."
d2 = "NLU – czyli rozumienie języka naturalnego, odpowiada za kolejny krok interakcji człowieka z maszyną. To dzięki NLU komputer nie tylko wie, jakie słowa wprowadził użytkownik, ale też jakie było ich znaczenie i kontekst. "
d3 = "NLG – czyli tworzenie naturalnego języka. Na podstawie zapytania i danych, które ma bot, odpowiedź powinna być przygotowana w sposób czytelny i jednoznaczny dla użytkownika."
d4 = "Polska i Węgry nie wyrażają zgody na mechanizm uzależnienia wypłaty unijnych funduszy od przestrzegania zasad praworządności. To komplikuje kwestie przyjęcia budżetu UE oraz Funduszu Odbudowy. "
```

```
# metoda normalizacyjna (sprowadzająca fragment tekstu do podstawowej formy słów)
def normalize(text):

    # tokenizujemy tekst
    words = word_tokenize(text)

    # zamieniamy na małe litery, odrzucamy liczby
    words = [word.lower() for word in words if word.isalpha()]

    # usuwamy słowa, które znajdują sie na liście stop_words
    words = [word for word in words if word not in stop_words]

    # każde słowo sprowadzamy do podstawowej formy
    words = [stempel_polimorf.stem(word) for word in words]

    # dodatkowo, czasami stemmery/lematyzery zwracają None dla zadanego słowa
    # więc takie przypadki trzeba odfiltrować
    words = [word for word in words if word != None]

    # tworzymy znormalizowany tekst
    return " ".join(words);
```

```
normalized_documents = [normalize(document) for document in [d1, d2, d3, d4]]
normalized_documents
```

```
from sklearn.feature_extraction.text import CountVectorizer

# tworzymy CV, czyli taką liczarkę, która zliczy nam wystepowanie słów w analizowanych
# przez nas dokumentach
# https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
cv = CountVectorizer()
count_matrix = cv.fit_transform(normalized_documents)

count_df = pd.DataFrame(count_matrix.toarray(), columns = cv.get_feature_names_out())
count_df
```

```
d1 = count_df.iloc[0].values
d2 = count_df.iloc[1].values
d3 = count_df.iloc[2].values
d4 = count_df.iloc[3].values
```

```
print(cosine(d1, d2))
print(cosine(d1, d3))
print(cosine(d1, d4))
```

#Analiza sentymentu

```
sentiment_df = pd.read_csv('dane_sentyment.csv', sep ='\t')
sentiment_df
```

```
sentiment_df['tekst znormalizowany'] = sentiment_df.apply(lambda row: normalize(row['tekst']), axis=1);
sentiment_df
```

```
from sklearn.feature_extraction.text import CountVectorizer

# podobnie jak w przypadku czterech zdań
# tworzymy CV, z tym, że ograniczamy wielkość naszego słownika do 1000 najczęściej występujących słów
cv = CountVectorizer(max_features = 5000)
count_matrix = cv.fit_transform(sentiment_df['tekst znormalizowany'] )

count_df = pd.DataFrame(count_matrix.toarray(), columns = cv.get_feature_names_out())
count_df
```

```
sentiment = sentiment_df['sentyment'].to_numpy()
sentiment
```

```
from sklearn.model_selection import train_test_split

# dobrą praktyką jest nie testowanie modelu na tych samych danych 
# na których je trenowaliśmy, dlatego skorzystamy z takiego mechanizmy
# który podzieli nam te dane w sposób losowy, w stosunku 70% (trening) : 30% (test)
X_train, X_test, y_train, y_test = train_test_split(count_df, sentiment, test_size = 0.3)
```

```
from sklearn.naive_bayes import MultinomialNB

# tworzymy model oparty o kalstyfikator bayesa (dla danych liczbowych - dyskretnych)
# https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html

sentiment_model = MultinomialNB().fit(X_train, y_train)

# sprawdzamy jaki procent trafilismy dobrze (na danych testowych)
predicted = sentiment_model.predict(X_test)
print(np.mean(predicted == y_test))
```

```
test_count_df = pd.DataFrame(
    cv.transform(['ten wykład to jakiś żart, kto go tu zaprosił']).toarray(), 
    columns = cv.get_feature_names_out()
)

predicted = sentiment_model.predict_proba(test_count_df)
predicted
```

```
test_count_df = pd.DataFrame(
    cv.transform(['wstęp o egzoplanetach mógł pominąć']).toarray(), 
    columns = cv.get_feature_names_out()
)

predicted = sentiment_model.predict_proba(test_count_df)
predicted
```

```
test_count_df = pd.DataFrame(
    cv.transform(['jestem zadowolony, bawię się naprawdę świetnie']).toarray(), 
    columns = cv.get_feature_names_out()
)

predicted = sentiment_model.predict_proba(test_count_df)
predicted
```

```
import os

# artykułów jest 10k, i nie ma sensu czekać, spróbujemy zrobić nasze testy
# na podstawie 300 pierwszych wylistowanych artykułów
for fn in os.listdir("artykuly10k")[:300]:
  f = open("artykuly10k/" + fn, "r")
  c = f.read()

  # dla każdego otwartego pliku, wyciągamy 'mięso' czyli artykuł dokumentu
  soup = BeautifulSoup(c)
  node = soup.find('div',{'class':'text-block'})

  if node != None:
    row = {'id':fn, 'tekst':node.text.replace("\n", " ")} 
    articles_df = articles_df.append(row, ignore_index = True)
```
